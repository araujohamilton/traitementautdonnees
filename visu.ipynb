- Fonction <b> scatter </b>: la fonction vous permet de créer un nuage de points
- La méthode <b> .rcParams </b> est utilisée pour agrandir la fenêtre graphique

```python
plt.rcParams["figure.figsize"]=[16,9]
plt.scatter(year_to_plot.index,year_to_plot['Temperature moyenne'], marker='x', color='red')
plt.show()
```

- pour utiliser l'option scatter color, python veut une liste d'entrée et non un dictionnaire
- option couleur: c = list ()

```python
plt.rcParams["figure.figsize"]=[16,9]
plt.scatter(year_to_plot.index,year_to_plot['Temperature moyenne'], c=[list(year_to_plot['Temperature moyenne'])],
            cmap="hot", 
            s=50)
plt.xlabel("Temps")
plt.ylabel("Température")
plt.title("Temperature")
plt.show()
```

- Nous avons utilisé la couleur de dispersion par défaut ()
- avec l'option <b> cmap </b>, nous pouvons choisir notre panel de couleurs: https://matplotlib.org/examples/color/colormaps_reference.html
- on choisira par exemple la palette de couleurs "sismique" via l'option <b> cmap </b>
- pour changer la forme et la taille des points: utilisez les options <b> marqueur </b> et <b> s </b>
- pour enregistrer un graphe: fonction pyplot: <b> savefig () </b>
- Pour ajouter une barre de couleur: fonction <b> colorbar () </b>
- Pour faire pivoter les étiquettes dans x: function <b> xticks () </b>

```python
plt.rcParams["figure.figsize"]=[16,9]
plt.scatter(year_to_plot.index,year_to_plot['Temperature moyenne'], c=list(year_to_plot['Temperature moyenne']),
            cmap="seismic",
            marker="D", 
            s=100)
plt.xlabel("Time")
plt.ylabel("Temperature")
plt.title("Temperature", y=1.05)
plt.colorbar()
plt.xticks(rotation=45)
plt.show()
#attention à l'emplacement
plt.savefig("/content/my_graph.png", bbox_inches="tight") 
```

##  Matplotlib subplot



Lors de la création d'un graphique, matplotlib:

- stocke un conteneur pour tous les graphiques
- stocke un conteneur pour que le graphique soit positionné sur une grille
- stocke les symboles visuels sur le graphique

Vous pouvez tracer différentes choses dans la même figure en utilisant la fonction de sous-tracé. Voici un exemple:

image.png

```python
fig = plt.figure()
ax1 = fig.add_subplot(2,2,1) # up and left 
ax2 = fig.add_subplot(2,2,2) # up and right 
ax3 = fig.add_subplot(2,2,3) # down and left 
ax4 = fig.add_subplot(2,2,4) # down and right
plt.show()
```

Voici une documentation sur le sous-intrigue:
https://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.subplot

```python
fig = plt.figure(figsize=(20,7))
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)
ax1.plot(dataframe_UQAM['2015'].index,year_to_plot['Temperature minimale'], color='lightblue') # je ne dis plus plt mais ax pour indiquer ce qui est présent
ax2.plot(dataframe_UQAM['2015'].index,year_to_plot['Temperature maximale'], color='darkgreen')
ax1.legend("min", loc='upper left')
ax2.legend("max", loc='upper left')
plt.show()
```

<h2>3- Présentation des principales représentations graphiques </h2>
<h3> a- Le bar chart </h3>
    
    - Exemple 1 : 

```python
data = pd.read_csv('data_visu2.csv')
data.head()
```

Répartition du nombre de personnes par class:

```python
passagersClasses = data['Pclass'].value_counts()
passagersClasses
```

```python
fig = plt.figure(figsize=(7,5))
ax1 = fig.add_subplot(111)
ax1.bar(range(len(passagersClasses)),passagersClasses)   # création du graphique
tick_positions = range(0,len(passagersClasses))          # définition de la position des échelles sur les axes
ax1.set_xticks(tick_positions)                           # ajout des échelles 
ax1.set_xticklabels(passagersClasses.index, rotation=45) # ajout des labels

plt.show()
```

    - Exemple 2 : 
    
   Répartition des classes avec un filtre

```python
passagersClassesWomen = data['Pclass'][data['Sex'] == 'female'].value_counts()
print(passagersClassesWomen)
passagersClassesMen = data['Pclass'][data['Sex'] == 'male'].value_counts()
print(passagersClassesMen)
```

```python
largeurBar = 0.25
position1W = np.arange(len(passagersClassesWomen))
position1M = [x + largeurBar for x in position1W]
print(position1W)
print(position1M)
```

```python
fig = plt.figure(figsize=(7,5))
ax1 = fig.add_subplot(111)

ax1.bar(position1W,passagersClassesWomen, width=largeurBar, label='Femmes')

ax1.bar(position1M,passagersClassesMen, width=largeurBar,label='Hommes')

tick_positions = range(0,len(passagersClasses))
ax1.set_xticks(tick_positions)
ax1.set_xticklabels(passagersClasses.index, rotation=45)

plt.legend()
plt.show()
```

    - Exemple 3 : 
   Création de bar empilé 

```python
passagersVieClasses = data['Pclass'][data['Survived'] == 1].value_counts()
print(passagersVieClasses)
passagersMortClasses = data['Pclass'][data['Survived'] == 0].value_counts()
print(passagersMortClasses)
```

```python
largeurBar = 0.5
position1 = np.arange(len(passagersVieClasses))

fig = plt.figure(figsize=(7,5))
ax1 = fig.add_subplot(111)

ax1.bar(position1,passagersVieClasses, width=largeurBar, label='Vie')

ax1.bar(position1,passagersMortClasses, width=largeurBar,label='décès',bottom=passagersVieClasses)

tick_positions = range(0,len(passagersClasses))
ax1.set_xticks(tick_positions)
ax1.set_xticklabels(passagersClasses.index, rotation=45)

plt.legend()
plt.show()
```

<h3> b- L' histogramme </h3>

```python
fig = plt.figure(figsize=(7,5))
ax1 = fig.add_subplot(111)

ax1.hist(data['Age'].dropna())          #on pousse la liste age en supprimant les valeurs manquantes
ax1.set_title("repartition de l'age")

plt.show()
```

```python
fig = plt.figure(figsize=(7,5))
ax1 = fig.add_subplot(111)

# on ajoute des filtres 
ageEnVie = data['Age'][data['Survived'] == 1].dropna()    
ageDeces = data['Age'][data['Survived'] == 0].dropna()
# on va pousser deux listes dans notre histogramme
ax1.hist([ageEnVie,ageDeces],bins=20,stacked=True,density = True)
ax1.set_title("Répartition de l'âge en fonction de la survie (en bleu) ou du décès (en orange)")
plt.show()
```

<h2>4- Aller plus loin avec Seaborn </h2>

Pour compléter et enrichir nos visualiations, une nouvelle bibliothèque Python sera utilisée: <b> Seaborn </b>.

- la bibliothèque Seaborn est basée sur matplotlib.
- il permet de dessiner des graphiques plus complexes

Pour plus d'informations:

https://jakevdp.github.io/PythonDataScienceHandbook/04.14-visualization-with-seaborn.html

Ce lien présente une galerie de types de graphiques à réaliser avec Seaborn:
https://seaborn.pydata.org/examples/index.html


image.png


https://seaborn.pydata.org/examples/index.html

```python
import seaborn as sns
sns.distplot(data["Age"].dropna())
sns.set()
plt.show()
```

```python
sns.kdeplot(data["Age"].dropna(), shade=True)
sns.set()
plt.xlabel("Age")
plt.show()
```

```python
sns.lmplot(x="Age", y="Fare", hue="Pclass", data=data,fit_reg=False)
plt.show()
```

```python
sns.jointplot("Age", "Fare", data=data, kind="reg")
plt.show()
```

```python
sns.jointplot("Age", "Fare", data=data, kind="hex")
plt.show()
```

```python
sns.boxplot(x=data["Age"])
plt.show()
```

image.png

```python
sns.boxplot(x="Pclass", y="Age", data=data)
plt.show()
```

```python
sns.violinplot(x="Pclass", y="Age", data=data)
plt.show()
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(12, 8)
sns.violinplot(x="Pclass", y="Age", data=data)
plt.show()
```
